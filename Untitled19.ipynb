```python
import matplotlib.pyplot as plt
import torch, torchvision

"""
im1 = torchvision.io.read_image("217730183_8f58409e7c_z.jpg").float()/255
im2 = torchvision.io.read_image("541870527_8fe599ec04_z.jpg").float()/255
im3 = torchvision.io.read_image("2124681469_7ee4868747_z.jpg",mode=torchvision.io.ImageReadMode.RGB).float()/255


im1 = torchvision.io.read_image("2711568708_89f2308b85_z.jpg").float()/255
im2 = torchvision.io.read_image("2928196999_acd5471d23_z.jpg").float()/255
im3 = torchvision.io.read_image("3016145160_497da1b387_z.jpg").float()/255
"""

im1 = torchvision.io.read_image("4683642953_2eeda0820e_z.jpg").float()/255
im2 = torchvision.io.read_image("6911037487_cc68a9d5a4_z.jpg").float()/255
im3 = torchvision.io.read_image("8139728801_60c233660e_z.jpg").float()/255


im1 = torch.nn.functional.interpolate(im1.unsqueeze(0), size=512)[0]
im2 = torch.nn.functional.interpolate(im2.unsqueeze(0), size=512)[0]
im3 = torch.nn.functional.interpolate(im3.unsqueeze(0), size=512)[0]

import matplotlib.pyplot as plt

visu = torch.cat([im1,im2,im3],dim=-1)
visu = visu.cpu().numpy().transpose(1,2,0)
plt.imshow(visu)
plt.axis('off')
plt.show()
```

maintenant utilisons sur ces images un algorithme de segmentation sémantique disponible dans le catalogue torchvision :

```python
net = torchvision.models.segmentation.deeplabv3_resnet50(weights=torchvision.models.segmentation.DeepLabV3_ResNet50_Weights.COCO_WITH_VOC_LABELS_V1)
#['__background__', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus', 'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike', 'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tvmonitor']
```

```python
with torch.no_grad():
  x = torch.stack([im1,im2,im3],dim=0)
  x = (x-0.5)/0.25 # preprocessing specifique à ce modèle ==> il faut regarder la doc https://pytorch.org/vision/main/models/generated/torchvision.models.segmentation.deeplabv3_resnet50.html#torchvision.models.segmentation.DeepLabV3_ResNet50_Weights
  z = net(x)["out"] # on prédit des cartes de score de confiance
  z = z[:,[0,8,12,15],:,:] # we keep only person, cat and dog class
  _,z = z.max(1) # on prend le meilleur score

# visualisation des prédictions : il faut transformer les indices de classes en couleur
couleur = torch.zeros(3,3,512,512)
couleur[:,0,:,:] = (z==1).float() # red for cat
couleur[:,1,:,:] = (z==2).float() # green for dog
couleur[:,2,:,:] = (z==3).float() # blue for person

visu = torch.cat([im1,im2,im3],dim=-1)
visubis = torch.cat([couleur[0],couleur[1],couleur[2]],dim=-1)
visu = torch.cat([visu,visubis],dim=1)
visu = visu.cpu().numpy().transpose(1,2,0)
plt.imshow(visu)
plt.axis('off')
plt.show()
```
