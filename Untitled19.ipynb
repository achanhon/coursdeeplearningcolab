```python
import matplotlib.pyplot as plt
import torch, torchvision

im1 = torchvision.io.read_image("176597054_281f51ee6b_z.jpg").float()/255
im2 = torchvision.io.read_image("4914789333_f5e52317af_z.jpg").float()/255
im3 = torchvision.io.read_image("2928196999_acd5471d23_z.jpg").float()/255

im1 = torch.nn.functional.interpolate(im1.unsqueeze(0), size=256)[0]
im2 = torch.nn.functional.interpolate(im2.unsqueeze(0), size=256)[0]
im3 = torch.nn.functional.interpolate(im3.unsqueeze(0), size=256)[0]

import matplotlib.pyplot as plt

visu = torch.cat([im1,im2,im3],dim=-1)
visu = visu.cpu().numpy().transpose(1,2,0)
plt.imshow(visu)
plt.axis('off')
plt.show()
```

maintenant utilisons sur ces images un algorithme de segmentation sémantique disponible dans le catalogue torchvision :
