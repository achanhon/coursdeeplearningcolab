```python
import matplotlib.pyplot as plt
import torch, torchvision

im1 = torchvision.io.read_image("176597054_281f51ee6b_z.jpg").float()/255
im2 = torchvision.io.read_image("4914789333_f5e52317af_z.jpg").float()/255
im3 = torchvision.io.read_image("2928196999_acd5471d23_z.jpg").float()/255

im1 = torch.nn.functional.interpolate(im1.unsqueeze(0), size=256)[0]
im2 = torch.nn.functional.interpolate(im2.unsqueeze(0), size=256)[0]
im3 = torch.nn.functional.interpolate(im3.unsqueeze(0), size=256)[0]

import matplotlib.pyplot as plt

visu = torch.cat([im1,im2,im3],dim=-1)
visu = visu.cpu().numpy().transpose(1,2,0)
plt.imshow(visu)
plt.axis('off')
plt.show()
```

maintenant utilisons sur ces images un algorithme de segmentation sémantique disponible dans le catalogue torchvision :

```python
net = torchvision.models.segmentation.deeplabv3_resnet50(weights=torchvision.models.segmentation.DeepLabV3_ResNet50_Weights.COCO_WITH_VOC_LABELS_V1)
```

```python
with torch.no_grad():
  x = torch.stack([im1,im2,im3],dim=0)
  x = (x-0.5)/0.25 # preprocessing specifique à ce modèle ==> il faut regarder la doc https://pytorch.org/vision/main/models/generated/torchvision.models.segmentation.deeplabv3_resnet50.html#torchvision.models.segmentation.DeepLabV3_ResNet50_Weights
  z = net(x)["out"] # on prédit des cartes de score de confiance
  _,z = z.max(1) # on prend le meilleur score

# visualisation des prédictions : il faut transformer les indices de classes en couleur
couleur = torch.zeros(3,3,256,256)
for i in range(3):
  for j in range(3):
      couleur[i][j] = ((z[i].long()*10+j*80)%255).float()/255
visu = torch.cat([im1,im2,im3],dim=-1)
visubis = torch.cat([couleur[0],couleur[1],couleur[2]],dim=-1)
visu = torch.cat([visu,visubis],dim=1)
visu = visu.cpu().numpy().transpose(1,2,0)
plt.imshow(visu)
plt.axis('off')
plt.show()
```
